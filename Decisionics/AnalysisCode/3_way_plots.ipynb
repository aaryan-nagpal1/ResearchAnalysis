```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the input file path
input_file_path = 'zupdate_2024_07_19 1-min 45 ml_b2.txt'

# Load the TXT file with comma-separated values
data = pd.read_csv(input_file_path, sep=',', header=None, names=['Time', 'Avg Pressure', 'Unused', 'Volume'])

# Extract the relevant columns
x = data['Time']
y = data['Avg Pressure']
z = data['Volume']

# Create a 3D plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.plot(x, y, z, label='Pressure and Volume Over Time')
ax.set_xlabel('Time (in seconds)')
ax.set_ylabel('Avg Pressure (in kPa)')
ax.set_zlabel('Volume (in ml)')
ax.set_title('Pressure and Volume Over Time')
ax.legend()

# Adjust layout
plt.tight_layout()

# Manually adjust margins
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# Show the plot
plt.show()
```
